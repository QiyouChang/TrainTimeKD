# Model A Graphs

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

chosen_teacher10 = 'CIFAR10-simplevit-large-32-Adam'
chosen_teacher100 = 'CIFAR100-simplevit-large-32-Adam'

# Create a figure and a set of subplots with 1 row and 2 columns
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# Define a function to highlight and annotate the chosen teacher
def highlight_teacher(ax, data, chosen_teacher):
    # Filter the specific teacher data point
    teacher_data = data[data['Name'] == chosen_teacher]
    # If the teacher is in the dataset, plot it
    if not teacher_data.empty:
        # Scatter the chosen teacher with red border
        ax.scatter(
            teacher_data['model.parameters']/1000000, 
            teacher_data['test_acc']*100, 
            s=100,  # Size of the marker
            facecolors='none',  # No fill color
            #edgecolors='red',  # Red border color
            linewidth=2,  # Width of the border
            label='Chosen'
        )
        # Annotate the chosen teacher point
        for _, row in teacher_data.iterrows():
            ax.annotate(
                'Chosen',
                xy=(row['model.parameters']/1000000, row['test_acc']*100),
                xytext=(20, 20),  # Position of text relative to point
                textcoords='offset points',  # Offset in points
                arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.2', color='gray'),  # Arrow properties
                bbox=dict(boxstyle='round,pad=0.5', facecolor='white', edgecolor='gray', alpha=0.5)  # Text box properties
            )

# Scatter plot for CIFAR-10
scatter1 = axs[0].scatter(
    Acifar10['model.parameters']/1000000, 
    Acifar10['test_acc']*100, 
    c=Acifar10['Relative Time (Process)']/(60*60),  # Color coding
    cmap='viridis',
    label='CIFAR-10'
)
highlight_teacher(axs[0], Acifar10, chosen_teacher10)

# Set the labels and title for the first plot
axs[0].set_xlabel('Model Parameters (Millions)')
axs[0].set_ylabel('Test Accuracy (%)')
axs[0].set_title('Model A runs on CIFAR-10')
#axs[0].set_ylim(80, 90)
# Add a color bar to the first plot
cbar1 = fig.colorbar(scatter1, ax=axs[0])
cbar1.set_label('Training Time (Hrs)')

# Scatter plot for CIFAR-100
scatter2 = axs[1].scatter(
    Acifar100['model.parameters']/1000000, 
    Acifar100['test_acc']*100, 
    c=Acifar100['Relative Time (Process)']/(60*60),  # Color coding
    cmap='viridis',
    label='CIFAR-100'
)
highlight_teacher(axs[1], Acifar100, chosen_teacher100)

# Set the labels and title for the second plot
axs[1].set_xlabel('Model Parameters (Millions)')
axs[1].set_ylabel('Test Accuracy (%)')
axs[1].set_title('Model A runs on CIFAR-100')
#axs[1].set_ylim(60, 70)
# Add a color bar to the second plot
cbar2 = fig.colorbar(scatter2, ax=axs[1])
cbar2.set_label('Training Time (Hrs)')

# Call highlight_teacher for each subplot
highlight_teacher(axs[0], Acifar10, chosen_teacher10)
highlight_teacher(axs[1], Acifar100, chosen_teacher100)

# Adjust the layout so there is no overlap and save the figure
plt.tight_layout()
plt.show()
```

```python
# summary table for CIFAR-10
import pandas as pd
import re

# Creating DataFrame
df = pd.DataFrame(Bcifar10)

# Filter out columns that end with '_MIN' or '_MAX'
filtered_df = df[[col for col in df.columns if not col.endswith(('_MIN', '_MAX'))]]

# Get the last non-null value from each remaining column
last_non_null_values = filtered_df.apply(lambda x: x.dropna().iloc[-1])

# Convert to DataFrame for better visualization and sort by value
result_df = pd.DataFrame(last_non_null_values, columns=['Last Non-Null Value']).reset_index()
result_df.rename(columns={'index': 'Column Name'}, inplace=True)

# Order the DataFrame by the 'Last Non-Null Value' in descending order
result_df = result_df.sort_values(by='Last Non-Null Value', ascending=False)

# Function to split column name correctly, handling scientific notations properly
def split_column_name(name):
    # Pattern to identify scientific notation
    pattern = re.compile(r'\d+e[-+]\d+', re.IGNORECASE)
    # Find all occurrences of scientific notation
    matches = pattern.findall(name)
    # Replace scientific notation with a placeholder to prevent splitting
    for match in matches:
        name = name.replace(match, match.replace('-', 'PLACEHOLDER'))
    # Split on hyphen
    parts = name.split('-')
    # Restore the original scientific notation
    parts = [part.replace('PLACEHOLDER', '-') for part in parts]
    return parts

# Splitting the column name into components without disturbing scientific notations
result_df['Components'] = result_df['Column Name'].apply(split_column_name)

print(result_df)
```
